```python
from collections import deque


def sliding_window_minimum(a, k):
    queue = deque()
    answer = []

    for i in range(len(a)):
        while queue and queue[0] <= i - k:
            queue.popleft()

        while queue and a[queue[-1]] >= a[i]:
            queue.pop()

        queue.append(i)

        if i >= k - 1:
            answer.append(a[queue[0]])

    return answer
```

```python
a = [11, 18, 14, 3, 5, 17, 4, 15, 6, 17, 25, 3, 14, 1, 5, 47, 6, 13, 20, 23, 6, 13]
k = 7

print(f'Массив: {a}')
print(f'Длина окна K: {k}')
```

```python
result = sliding_window_minimum(a, k)

print('Минимумы в окнах:')
print(result)
```

```python
print('Подробно по каждому окну:')

for i in range(len(a) - k + 1):
    window = a[i:i + k]
    print(f'{i + 1}-{i + k}: {window} -> минимум {min(window)}')
```
