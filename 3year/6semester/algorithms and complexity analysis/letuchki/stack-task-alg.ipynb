```python
def nearest_greater_right_indices(arr):
    n = len(arr)
    answer = [-1] * n
    stack = []

    for i in range(n - 1, -1, -1):
        while stack and arr[stack[-1]] <= arr[i]:
            stack.pop()

        if stack:
            answer[i] = stack[-1] + 1

        stack.append(i)

    return answer
```

```python
a = [14, 20, 21, 13, 5, 87, 9, 61, 3, 45, 10, 36]

print(f'Массив: {a}')
print(f'Индексы элементов (с 1): {list(range(1, len(a) + 1))}')
```

```python
result = nearest_greater_right_indices(a)
print(f'Ближайший справа больший индекс: {result}')
```
