```python
def max_non_overlapping_intervals(intervals):
    intervals.sort(key=lambda x: x[1])
    
    result = []
    if not intervals:
        return result
    
    last_r = intervals[0][1]
    result.append(intervals[0])
    
    for i in range(1, len(intervals)):
        if intervals[i][0] >= last_r:
            result.append(intervals[i])
            last_r = intervals[i][1]
            
    return result
```

```python
data = [(1, 4), (3, 5), (0, 6), (5, 7), (3, 8), (5, 9), (6, 10), (8, 11), (2, 13), (12, 14)]
print("Максимальное подмножество:", max_non_overlapping_intervals(data))
```

```python
def max_applications(apps):
    apps.sort(key=lambda x: x[1])
    
    selected = []
    last_end_time = -1
    
    for start, end in apps:
        if start >= last_end_time:
            selected.append((start, end))
            last_end_time = end
            
    return selected
```

```python
applications = [(9, 11), (10, 12), (11, 13), (12, 14), (13, 15)]
print("Выбранные заявки:", max_applications(applications))
```

```python
def min_interval_cover(intervals, target_L, target_R):
    intervals.sort()
    
    selected = []
    current_right = target_L
    i = 0
    n = len(intervals)
    
    while current_right < target_R:
        best_r = current_right
        best_interval = None
        
        while i < n and intervals[i][0] <= current_right:
            if intervals[i][1] > best_r:
                best_r = intervals[i][1]
                best_interval = intervals[i]
            i += 1
        
        if best_interval is None:
            return "Покрытие невозможно"
        
        selected.append(best_interval)
        current_right = best_r
        
    return selected
```

```python
test_intervals = [(0, 4), (2, 7), (5, 10), (8, 12)]
print("Минимальное покрытие [0, 10]:", min_interval_cover(test_intervals, 0, 10))
```
