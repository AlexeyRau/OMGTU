```python
import itertools
import time

class Item:
    def __init__(self, name, weight, value):
        self.name = name
        self.weight = weight
        self.value = value
        self.ratio = value / weight
```

```python
def brute_force_knapsack(items, capacity):
    best_value = -1
    best_combination = []
    
    n = len(items)
    for i in range(n + 1):
        for combination in itertools.combinations(items, i):
            total_weight = sum(item.weight for item in combination)
            total_value = sum(item.value for item in combination)
            
            if total_weight <= capacity and total_value > best_value:
                best_value = total_value
                best_combination = combination
                
    return best_combination, best_value
```

```python
def greedy_knapsack(items, capacity):
    sorted_items = sorted(items, key=lambda x: x.ratio, reverse=True)
    
    total_value = 0
    total_weight = 0
    knapsack = []
    
    for item in sorted_items:
        if total_weight + item.weight <= capacity:
            knapsack.append(item)
            total_weight += item.weight
            total_value += item.value
            
    return knapsack, total_value
```

```python
items_list = [
    Item("Золотой слиток", 10, 60),
    Item("Картина", 20, 100),
    Item("Ваза", 30, 120),
    Item("Часы", 5, 50),
    Item("Ноутбук", 15, 90)
]
W = 35

print(f"Грузоподъемность рюкзака: {W}\n")

start = time.time()
bf_items, bf_val = brute_force_knapsack(items_list, W)
end = time.time()
print("Полный перебор")
print(f"Предметы: {[item.name for item in bf_items]}")
print(f"Максимальная стоимость: {bf_val}")
print(f"Время выполнения: {end - start:.6f} сек.\n")

start = time.time()
gr_items, gr_val = greedy_knapsack(items_list, W)
end = time.time()
print("Жадный алгоритм")
print(f"Предметы: {[item.name for item in gr_items]}")
print(f"Суммарная стоимость: {gr_val}")
print(f"Время выполнения: {end - start:.6f} сек.")
```
