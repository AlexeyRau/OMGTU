```python
def solve_task_1(jobs):
   
    max_deadline = max(job[1] for job in jobs)
    schedule = [None] * (max_deadline + 1)
    
    jobs.sort(key=lambda x: x[2], reverse=True)
    
    total_profit = 0
    selected_jobs = []

    for job in jobs:
        name, deadline, price = job
        for day in range(deadline, 0, -1):
            if schedule[day] is None:
                schedule[day] = name
                total_profit += price
                selected_jobs.append(name)
                break
                
    return selected_jobs, total_profit

data1 = [('A', 2, 40), ('B', 1, 25), ('C', 2, 30), ('D', 1, 15), ('E', 3, 20)]
jobs_res, profit = solve_task_1(data1)
print(f"Выбранные заказы: {jobs_res}, Общая стоимость: {profit}")
```

```python
def solve_task_2(children):
    if not children:
        return 0, []

    children.sort()
    
    groups = []
    if children:
        current_group = [children[0]]
        for i in range(1, len(children)):
            if children[i] - current_group[0] <= 2:
                current_group.append(children[i])
            else:
                groups.append(current_group)
                current_group = [children[i]]
        groups.append(current_group)
        
    return len(groups), groups

kids_ages = [5, 3, 8, 4, 10, 2, 7]
count, distribution = solve_task_2(kids_ages)
print(f"Количество групп: {count}, Распределение: {distribution}")
```
