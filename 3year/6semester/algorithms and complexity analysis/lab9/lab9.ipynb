```python
def greedy_graph_coloring(jobs, mechanisms_map):
    adj = {job: set() for job in jobs}
    for i in range(len(jobs)):
        for j in range(i + 1, len(jobs)):
            job1, job2 = jobs[i], jobs[j]
            if mechanisms_map[job1] & mechanisms_map[job2]:
                adj[job1].add(job2)
                adj[job2].add(job1)

    sorted_jobs = sorted(jobs, key=lambda x: len(adj[x]), reverse=True)

    job_colors = {}
    current_color = 1

    while len(job_colors) < len(jobs):
        for job in sorted_jobs:
            if job not in job_colors:
                can_color = True
                for neighbor in adj[job]:
                    if job_colors.get(neighbor) == current_color:
                        can_color = False
                        break
                
                if can_color:
                    job_colors[job] = current_color
        
        current_color += 1

    return job_colors, current_color - 1

jobs_list = [f"a{i}" for i in range(1, 11)]
data = {
    "a1": {"b1", "b5", "b8"},
    "a2": {"b2", "b3"},
    "a3": {"b2", "b7", "b9"},
    "a4": {"b1", "b6", "b7", "b9"},
    "a5": {"b3", "b5"},
    "a6": {"b1", "b4", "b7", "b9"},
    "a7": {"b2", "b9"},
    "a8": {"b3", "b6", "b8"},
    "a9": {"b2", "b4", "b5"},
    "a10": {"b1", "b6", "b9"}
}

result, total_time = greedy_graph_coloring(list(data.keys()), data)

print(f"Минимальное время выполнения: {total_time} ед.")
for time_slot in range(1, total_time + 1):
    current_jobs = [job for job, slot in result.items() if slot == time_slot]
    print(f"Таймслот {time_slot}: {', '.join(current_jobs)}")
```

```python
def greedy_cargo_placement(cargos, conflicts):
    adj = {cargo: set(conflicts.get(cargo, set())) for cargo in cargos}
    for cargo in cargos:
        for neighbor in list(adj[cargo]):
            adj[neighbor].add(cargo)  # делаем undirected
    
    sorted_cargos = sorted(cargos, key=lambda x: len(adj[x]), reverse=True)
    
    assignment = {}
    current_container = 1
    
    while len(assignment) < len(cargos):
        for cargo in sorted_cargos:
            if cargo not in assignment:
                can_place = True
                for neighbor in adj[cargo]:
                    if assignment.get(neighbor) == current_container:
                        can_place = False
                        break
                if can_place:
                    assignment[cargo] = current_container
        current_container += 1
    
    num_containers = current_container - 1
    return assignment, num_containers

cargos_list = ['g1', 'g2', 'g3', 'g4', 'g5', 'g6', 'g7', 'g8', 'g9', 'g10']

conflicts_data = {
    'g1': {'g2', 'g5', 'g8'},
    'g2': {'g1', 'g3', 'g6'},
    'g3': {'g2', 'g4', 'g7', 'g9'},
    'g4': {'g3', 'g5', 'g8'},
    'g5': {'g1', 'g4', 'g6', 'g10'},
    'g6': {'g2', 'g5', 'g7'},
    'g7': {'g3', 'g6', 'g9'},
    'g8': {'g1', 'g4'},
    'g9': {'g3', 'g7', 'g10'},
    'g10': {'g5', 'g9'}
}

assignment, num_containers = greedy_cargo_placement(cargos_list, conflicts_data)

print(f"Минимальное количество контейнеров (по жадному алгоритму): {num_containers}\n")
for container in range(1, num_containers + 1):
    items = [cargo for cargo, cont in assignment.items() if cont == container]
    print(f"Контейнер {container}: {', '.join(items)}")
```
