```python
class Node:
    def __init__(self, char=None, freq=0):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None
```

```python
def build_frequency_dict(text):
    freq = {}
    for char in text:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq
```

```python
def build_huffman_tree(freq_dict):
    if not freq_dict:
        return None
    
    nodes = [Node(char, freq) for char, freq in freq_dict.items()]
    
    while len(nodes) > 1:
        nodes.sort(key=lambda x: x.freq)
        
        left = nodes.pop(0)
        right = nodes.pop(0)
        
        parent = Node(None, left.freq + right.freq)
        parent.left = left
        parent.right = right
        
        nodes.append(parent)
    
    return nodes[0]
```

```python
def build_codes(node, current_code="", codes=None):
    if codes is None:
        codes = {}
    if node is None:
        return codes
    
    if node.char is not None:
        codes[node.char] = current_code
        return codes
    
    build_codes(node.left, current_code + "0", codes)
    build_codes(node.right, current_code + "1", codes)
    return codes
```

```python
def huffman_encode(text):
    if not text:
        return {}, ""
    
    freq_dict = build_frequency_dict(text)
    tree = build_huffman_tree(freq_dict)
    codes = build_codes(tree)
    
    encoded = ''.join(codes[char] for char in text)
    
    return codes, encoded
```

```python
def huffman_decode(encoded_text, codes):
    if not encoded_text:
        return ""
    
    reverse_codes = {code: char for char, code in codes.items()}
    
    decoded = []
    current = ""
    
    for bit in encoded_text:
        current += bit
        if current in reverse_codes:
            decoded.append(reverse_codes[current])
            current = ""
    
    return ''.join(decoded)
```

```python
if __name__ == "__main__":
    text = input("Введите текст для кодирования: ")
    
    print("\nОригинальный текст:", text)
    print("Длина:", len(text), "символов")
    
    codes, encoded = huffman_encode(text)
    
    print("\n=== Словарь кодов ===")
    for char in sorted(codes.keys()):
        print(f"'{char}' : {codes[char]}")
    
    print("\nЗакодированная строка:")
    print(encoded)
    print("Длина закодированной строки:", len(encoded), "бит")
    
    decoded = huffman_decode(encoded, codes)
    
    print("\nДекодированный текст:", decoded)
    print("Восстановление успешно:", text == decoded)
    
    original_bits = len(text) * 8
    print(f"\nИсходный размер: {original_bits} бит")
    print(f"Сжатый размер:   {len(encoded)} бит")
    print(f"Сжатие: {len(encoded)/original_bits*100:.1f}% от исходного")
```
