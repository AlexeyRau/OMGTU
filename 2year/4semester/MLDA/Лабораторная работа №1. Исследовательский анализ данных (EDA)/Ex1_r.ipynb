<font size=4><b>Предположения для заполнения пропущенного значения:</b>

- Model: Заполнить пропущенные значения на основе наиболее часто встречающихся моделей для каждого производителя
- Year: Заполнить пропущенные значения медианным годом выпуска для каждой модели или производителя.
- Transmission: Заполнить пропуски наиболее часто встречающимся типом трансмиссии для каждой модели или производителя.
- Distance: Заполнить пропущенные значения медианным пробегом для автомобилей с аналогичными характеристиками.
- Style: Заполнить пропущенные значения наиболее часто встречающимся стилем кузова для каждой модели или производителя.


```python
def improved_fill_missing(df):
    df_filled = df.copy()
    
    # 1. Заполнение Model на основе Make
    df_filled['Model'] = df_filled.groupby('Make')['Model'].transform(
        lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else 'Unknown'
    )
    
    # 2. Заполнение Year на основе Model и Make
    year_median = df_filled.groupby(['Make', 'Model'])['Year'].transform('median')
    df_filled['Year'] = df_filled['Year'].fillna(year_median)
    # Если остались пропуски, заполняем общим медианным годом
    df_filled['Year'] = df_filled['Year'].fillna(df_filled['Year'].median())
    
    # 3. Заполнение Transmission
    df_filled['Transmission'] = df_filled.groupby(['Make', 'Model'])['Transmission'].transform(
        lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else 'Manual'
    )
    
    # 4. Заполнение Distance
    distance_median = df_filled.groupby(['Make', 'Model', 'Year'])['Distance'].transform('median')
    df_filled['Distance'] = df_filled['Distance'].fillna(distance_median)
    df_filled['Distance'] = df_filled['Distance'].fillna(df_filled['Distance'].median())
    
    # 5. Заполнение Style
    df_filled['Style'] = df_filled.groupby(['Make', 'Model'])['Style'].transform(
        lambda x: x.fillna(x.mode()[0]) if not x.mode().empty else 'Sedan'
    )
    
    return df_filled

# Применяем улучшенное заполнение
data = improved_fill_missing(data)
```

```python
f = lambda x: x.mode().iat[0] if x.notna().any() else np.nan
group_model = data.groupby('Make')['Model'].transform(f)
data['Model'] = data['Model'].fillna(group_model)
(data.isna().sum()/data.shape[0]*100).sort_values(ascending=False)
```

```python
group_year = data.groupby('Model')['Year'].transform('median')
data['Year'] = data['Year'].fillna(group_year)
(data.isna().sum()/data.shape[0]*100).sort_values(ascending=False)
```

```python
group_transmission = data.groupby('Model')['Transmission'].transform(f)
data ['Transmission'] = data ['Transmission'].fillna(group_transmission)
if data['Transmission'].isna().any():
    group_transmission_make = data.groupby('Make')['Transmission'].transform(f)
    data['Transmission'] = data['Transmission'].fillna(group_transmission_make)

(data.isna().sum()/data.shape[0]*100).sort_values(ascending=False)
```

```python
group_distance = data.groupby('Model')['Distance'].transform('median')
data['Distance'] = data['Distance'].fillna(group_distance)
(data.isna().sum()/data.shape[0]*100).sort_values(ascending=False)
```

```python
group_style = data.groupby('Model')['Style'].transform(f)
data['Style'] = data['Style'].fillna(group_style)
(data.isna().sum()/data.shape[0]*100).sort_values(ascending=False)
```

## Изучение и обработка выбросов

```python
outlier = data[columns]
outlier
```

```python
# Определить нижние пределы для выбросов
Q1 = outlier.quantile(0.25)
Q1
```

```python
# Определить верхние пределы выбросов
Q3 = outlier.quantile(0.75)
Q3
```

```python
def remove_outliers_iqr(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Для Distance и Price устанавливаем разумные минимальные значения
        if col == 'Distance':
            lower_bound = max(lower_bound, 1000)  # Минимальный пробег 1000 км
        elif col == 'Price(euro)':
            lower_bound = max(lower_bound, 500)   # Минимальная цена 500 евро
        elif col == 'Year':
            lower_bound = max(lower_bound, 1980)  # Минимальный год 1980
        elif col == 'Engine_capacity(cm3)':
            lower_bound = max(lower_bound, 800)   # Минимальный объем 800 см3
            
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    
    return df_clean
```

```python
numeric_columns = ['Year', 'Distance', 'Engine_capacity(cm3)', 'Price(euro)']
data_filtered = remove_outliers_iqr(data, numeric_columns)
```

```python
from sklearn.preprocessing import LabelEncoder
data_filtered = data[data.index.isin(index_list)]

from sklearn.preprocessing import LabelEncoder
import pandas as pd

def safe_label_encode(df, columns):
    df_encoded = df.copy()
    label_encoders = {}
    
    for col in columns:
        # Проверяем, есть ли пропуски
        if df_encoded[col].isna().any():
            print(f"Внимание: в колонке {col} есть пропуски перед кодированием")
            # Заполняем наиболее частым значением
            most_frequent = df_encoded[col].mode()[0]
            df_encoded[col] = df_encoded[col].fillna(most_frequent)
        
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = le
        
        print(f"Закодировано {col}: {len(le.classes_)} уникальных значений")
    
    return df_encoded, label_encoders

categories = ['Make', 'Model', 'Style', 'Fuel_type', 'Transmission']
data_encoded, label_encoders = safe_label_encode(data_filtered, categories)

#data_encoded = data_encoded.drop(['Vehicle_age', 'Avg_mileage'], axis=1)
data_encoded.to_csv('A:/OMGTU/OMGTU/2year/4semester/MLDA/datasets/filtered_moldova_cars_task.csv', index=False)
data_encoded
```

```python
def check_data_quality(df):
    print("=== КАЧЕСТВО ДАННЫХ ===")
    print(f"Размер датасета: {df.shape}")
    print(f"Пропуски: {df.isna().sum().sum()}")
    print(f"Дубликаты: {df.duplicated().sum()}")
    
    print("\n=== СТАТИСТИКИ ЧИСЛОВЫХ КОЛОНОК ===")
    numeric_cols = ['Year', 'Distance', 'Engine_capacity(cm3)', 'Price(euro)']
    print(df[numeric_cols].describe())
    
    print("\n=== ПРОВЕРКА АНОМАЛИЙ ===")
    for col in numeric_cols:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        outliers = df[(df[col] < lower) | (df[col] > upper)].shape[0]
        print(f"{col}: {outliers} выбросов ({outliers/len(df)*100:.2f}%)")

# Проверяем качество
check_data_quality(data_encoded)
```
