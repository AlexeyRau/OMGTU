<font size=4>
Датасет содержит следующие поля:

- `Diabetes_012` — наличие диабета (0 = нет/только при беременности, 1 = преддиабет, 2 = диабет)
- `HighBP` — диагностировано высокое кровяное давление 
- `HighChol` — высокий уровень холестерина
- `CholCheck` — проверка холестерина за последние 5 лет
- `BMI` — индекс массы тела
- `Smoker` — курение более 100 сигарет за жизнь 
- `Stroke` — перенесенный инсульт 
- `HeartDiseaseorAttack` — ишемическая болезнь сердца или инфаркт 
- `PhysActivity` — физическая активность вне работы 
- `Fruits` — регулярное потребление фруктов 
- `Veggies` — регулярное потребление овощей 
- `HvyAlcoholConsump` — тяжелое употребление алкоголя 
- `AnyHealthcare` — наличие медицинской страховки 
- `NoDocbcCost` — отказ от визита к врачу из-за стоимости 
- `GenHlth` — самооценка общего здоровья 
- `MentHlth` — количество дней с плохим психическим здоровьем за месяц 
- `PhysHlth` — количество дней с плохим физическим здоровьем за месяц 
- `DiffWalk` — трудности при ходьбе или подъеме по лестнице 
- `Sex` — пол (0 = женский, 1 = мужской)
- `Age` — возрастная категория 
- `Education` — уровень образования 
- `Income` — уровень дохода

## Коррекция типов данных

```python
data.info()
```

```python
columns_to_convert = [
    'Diabetes_012', 'HighBP', 'HighChol', 'CholCheck', 'BMI', 
    'Smoker', 'Stroke', 'HeartDiseaseorAttack', 'PhysActivity',
    'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare',
    'NoDocbcCost', 'GenHlth', 'MentHlth', 'PhysHlth', 'DiffWalk',
    'Sex', 'Age', 'Education', 'Income'
]

data[columns_to_convert] = data[columns_to_convert].astype('int')
data.info()
```

## Обогащение данных

```python
age_mapping = {
    1: '18-24',
    2: '25-29',
    3: '30-34',
    4: '35-39',
    5: '40-44',
    6: '45-49',
    7: '50-54',
    8: '55-59',
    9: '60-64',
    10: '65-69',
    11: '70-74',
    12: '75-79',
    13: '80+'
}

data['age_group'] = data['Age'].map(age_mapping)
data['age_group'].value_counts()
```

```python
bmi_bins = [0, 16.5, 18.5, 25, 30, 35, 40, 100]
bmi_labels = [
    'Severe underweight',
    'Underweight',
    'Normal weight',
    'Overweight',
    'Obesity class I',
    'Obesity class II',
    'Obesity class III'
]
data['bmi_category'] = pd.cut(data['BMI'], bins=bmi_bins, labels=bmi_labels)
data['bmi_category'].value_counts()
```

## Факторы риска диабета

```python
group_9 = data.query('Age == 9')
group_10 = data.query('Age == 10')
```

```python
sns.lmplot(x="BMI", y="Diabetes_012", data=group_9)
plt.title('60-64 лет')
sns.lmplot(x="BMI", y="Diabetes_012", data=group_10)
plt.title('65-69 года')
```

```python
# вычисление корелляции
BMI_group_9 = group_9[['BMI','Diabetes_012']]
corr_group_9 = BMI_group_9.corr()
BMI_group_10 = group_10[['BMI','Diabetes_012']]
corr_group_10 = BMI_group_10.corr()
```

```python
round(corr_group_9,3)
```

```python
round(corr_group_10,3)
```

**Заключение** :
<br>
рост BMI для обоих возрастных групп увеличивает риск диабета. У старшей возрастной группы это более выражено

```python
sns.lmplot(x="Income", y="Diabetes_012", data=group_9, height = 5, aspect =3)
plt.title('9')
sns.lmplot(x="Income", y="Diabetes_012", data=group_10, height = 5, aspect =3)
plt.title('10')
```

```python
Income_9 = group_9[['Income','Diabetes_012']]
Income_9_corr = Income_9.corr()
Income_10 = group_10[['Income','Diabetes_012']]
Income_10_corr = Income_10.corr()
```

```python
highbp_diabetes_9 = pd.crosstab(group_9['HighBP'], group_9['Diabetes_012'], normalize='index') * 100
highbp_diabetes_10 = pd.crosstab(group_10['HighBP'], group_10['Diabetes_012'], normalize='index') * 100

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
highbp_diabetes_9.plot(kind='bar', stacked=True, ax=axes[0], title='60-64 года (HighBP)', grid=True)
highbp_diabetes_10.plot(kind='bar', stacked=True, ax=axes[1], title='65-69 лет (HighBP)', grid=True)
plt.ylabel('Доля пациентов (%)')
plt.show()
```

```python
highchol_diabetes_9 = pd.crosstab(group_9['HighChol'], group_9['Diabetes_012'], normalize='index')*100
highchol_diabetes_10 = pd.crosstab(group_10['HighChol'], group_10['Diabetes_012'], normalize='index')*100

fig, axes = plt.subplots(1, 2, figsize=(15,5))
highchol_diabetes_9.plot(kind='bar', stacked=True, ax=axes[0], title='60-64 года (HighChol)', grid=True)
highchol_diabetes_10.plot(kind='bar', stacked=True, ax=axes[1], title='65-69 лет (HighChol)', grid=True)
plt.ylabel('Доля пациентов (%)')
plt.show()
```

```python
physact_diabetes_9 = pd.crosstab(group_9['PhysActivity'], group_9['Diabetes_012'], normalize='index')*100
physact_diabetes_10 = pd.crosstab(group_10['PhysActivity'], group_10['Diabetes_012'], normalize='index')*100

fig, axes = plt.subplots(1, 2, figsize=(15,5))
physact_diabetes_9.plot(kind='bar', stacked=True, ax=axes[0], title='60-64 года (PhysActivity)', grid=True)
physact_diabetes_10.plot(kind='bar', stacked=True, ax=axes[1], title='65-69 лет (PhysActivity)', grid=True)
plt.ylabel('Доля пациентов (%)')
plt.show()
```

**Заключение**:

- Оба графика показывают, что наличие высокого кровяного давления, холестерина увеличвает шанс наличия диабета (У группы 65-69 немного в большей степени).
- Наличие физической активности вне работы снижает риск диабета

```python
fig, axes = plt.subplots(2, 2, figsize=(15,10))

pd.crosstab(group_9['Fruits'], group_9['Diabetes_012'], normalize='index').plot(kind='bar', ax=axes[0,0], title='60-64 года (Fruits)', grid=True)
pd.crosstab(group_10['Fruits'], group_10['Diabetes_012'], normalize='index').plot(kind='bar', ax=axes[0,1], title='65-69 лет (Fruits)', grid=True)

pd.crosstab(group_9['Veggies'], group_9['Diabetes_012'], normalize='index').plot(kind='bar', ax=axes[1,0], title='60-64 года (Veggies)', grid=True)
pd.crosstab(group_10['Veggies'], group_10['Diabetes_012'], normalize='index').plot(kind='bar', ax=axes[1,1], title='65-69 лет (Veggies)', grid=True)

plt.tight_layout()
plt.show()
```

```python
diffwalk_diabetes_9 = pd.crosstab(group_9['DiffWalk'], group_9['Diabetes_012'], normalize='index')*100
diffwalk_diabetes_10 = pd.crosstab(group_10['DiffWalk'], group_10['Diabetes_012'], normalize='index')*100

fig, axes = plt.subplots(1, 2, figsize=(15,5))
diffwalk_diabetes_9.plot(kind='bar', stacked=True, ax=axes[0], title='60-64 года (DiffWalk)', grid=True)
diffwalk_diabetes_10.plot(kind='bar', stacked=True, ax=axes[1], title='65-69 лет (DiffWalk)', grid=True)
plt.ylabel('Доля пациентов (%)')
plt.show()
```

**Заключение:**

Наличие трудностей при ходьбе у обеих групп могут свидетельствовать о наличии диаебета

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Diabetes_012', y='GenHlth', data=group_9)
plt.title('60-64 года: Связь между GenHlth и Diabetes_012')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='Diabetes_012', y='GenHlth', data=group_10)
plt.title('65-69 лет: Связь между GenHlth и Diabetes_012')
plt.show()
```

```python
numeric_cols = ['HighBP', 'HighChol', 'BMI', 'Smoker', 'Stroke', 
               'HeartDiseaseorAttack', 'PhysActivity', 'GenHlth',
               'MentHlth', 'PhysHlth', 'Income']

def plot_corr(df, age_group):
    corr = df[numeric_cols + ['Diabetes_012']].corr()
    plt.figure(figsize=(12,8))
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm',
               annot_kws={"size": 10}, vmin=-1, vmax=1)
    plt.title(f'Корреляция факторов (Возраст {age_group})')
    plt.show()

plot_corr(group_9, '60-64 года')
plot_corr(group_10, '65-69 лет')
```
