# EDA

```python
from scipy.stats import pointbiserialr, spearmanr, chi2_contingency

# Пример для бисериальной корреляции
def calc_biserial(df, binary_col, target_col):
    return pointbiserialr(df[binary_col], df[target_col])

# Пример для Спирмена
def calc_spearman(df, ordinal_col, target_col):
    return spearmanr(df[ordinal_col], df[target_col])

# Пример для категориальных переменных
def calc_chi2(df, categorical_col, target_col):
    contingency_table = pd.crosstab(df[categorical_col], df[target_col])
    return chi2_contingency(contingency_table)
```

###  DiffWalk (бинарная) vs Diabetes_012:

```python
# Парные графики для DiffWalk
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)

# Общие настройки
plt.suptitle('Связь трудностей при ходьбе с диабетом', y=1.05, fontsize=14)
colors = ['#66c2a5', '#fc8d62', '#8da0cb']  # Цвета для 0,1,2 категорий диабета

for i, (age_group, data) in enumerate(zip(['60-64 года', '65-69 лет'], [group_9, group_10])):
    # Расчет корреляции
    r, p = pointbiserialr(data['DiffWalk'], data['Diabetes_012'])
    
    # Нормализованная таблица сопряженности
    ct = pd.crosstab(data['DiffWalk'], data['Diabetes_012'], normalize='index')
    
    # Построение stacked bar plot
    ct.plot(kind='bar', 
            stacked=True, 
            ax=axes[i],
            color=colors,
            width=0.7,
            edgecolor='black')
    
    # Аннотации
    axes[i].text(0.95, 0.95, 
                f'r = {r:.3f}\np = {p:.4f}', 
                ha='right', va='top',
                transform=axes[i].transAxes,
                bbox=dict(facecolor='white', alpha=0.8))
    
    # Настройка осей
    axes[i].set_title(age_group)
    axes[i].set_xticklabels(['Нет трудностей', 'Есть трудности'], rotation=0)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Доля пациентов' if i == 0 else '')
    axes[i].grid(axis='y', linestyle='--', alpha=0.7)

# Общая легенда
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, ['Нет диабета', 'Предиабет', 'Диабет 2 типа'],
           loc='upper right', 
           bbox_to_anchor=(0.92, 0.95),
           frameon=True)

plt.tight_layout()
plt.show()
```

**Заключение:**

Наличие трудностей при ходьбе у обеих групп могут свидетельствовать о наличии диаебета

Необходимо выявить "аномальные" данные, отсутствующие значения и некорректные типы данных.
